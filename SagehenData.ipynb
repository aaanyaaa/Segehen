```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_formats = {'svg',}

import sagehen_pandas as sage_pd
```

```python
# month_dict = {'1': 'JAN', '2':'FEB', '3':'MAR', '4':'APR', '5':'MAY', '6':'JUN', '7':'JUL', '8':'AUG', '9':'SEP', '10':'OCT', '11':'NOV', '12':'DEC'}
rainfall = sage_pd.year_search(1997,2017,['precip']).resample('M').sum()
rainfall['year'] = rainfall.index.year
rainfall['month'] = rainfall.index.month
rainfall = rainfall[rainfall['precip']<10]
rainfall.dropna()
rainfall_pivot = rainfall.pivot(index='month', columns='year', values='precip')
rainfall_pivot
```

```python
months = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
# rainfall_pivot.drop('2000', axis=1)
ax = rainfall_pivot.plot(
    title="Ave. Monthly Rainfall",
    colormap='nipy_spectral',
    figsize=(10,8),
)
ax.set_xticks(rainfall_pivot.index)
ax.set_xticklabels(months)
ax.set_xlabel("Months")
ax.set_ylabel("mm")
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), title="Years")
ax.grid(True)
ax.set_yscale('log')
```
